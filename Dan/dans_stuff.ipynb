```python
from math import sin, cos, sqrt, atan2, radians

def lat_lon_distance(lat_1, lon_1, lat_2, lon_2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat_1)
    lon1 = radians(lon_1)
    lat2 = radians(lat_2)
    lon2 = radians(lon_2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
clus_pairwise_distances = np.zeros((clustered_locs.shape[0], clustered_locs.shape[0]))
for i in range(clustered_locs.shape[0]):
    for j in range(clustered_locs.shape[0]):
        lat_i = clustered_locs.iloc[i]['lat'] 
        lon_i = clustered_locs.iloc[i]['lon'] 
        lat_j = clustered_locs.iloc[j]['lat'] 
        lon_j = clustered_locs.iloc[j]['lon'] 
        clus_pairwise_distances[i][j] = lat_lon_distance(lat_i, lon_i, lat_j, lon_j)
```

```python
class Cluster:
    def __init__(self, cluster_id, name, capacity, lat, lon):
        self.cluster_id = cluster_id
        self.name = name
        self.capacity = capacity
        self.lat = lat
        self.lon = lon
    
    num_bikes = 0
    
    def add_bike(self):
        if (self.num_bikes + 1) >= self.capacity: 
            raise(ValueError('too many bikes at cluster ' + str(self.cluster_id) + ': ' + self.name))
        else:
            self.num_bikes = self.num_bikes + 1
        
    def remove_bike(self):
        if (self.num_bikes - 1) == -1:
            raise(ValueError('cluster ' + str(self.cluster_id) + ': ' + self.name + ' is empty'))
    
    def is_close_to_dead(self):
        lower_bound = 0.1 * self.capacity
        upper_bound = 0.1 * self.capacity
        ratio = self.num_bikes / self.capacity
        if ratio > lower_bound and ratio < upper_bound:
            return False
        else:
            return True
```

```python
class MenWithVen:
        
    def __init__(self, ven_start_pos, current_time, cluster_pairwise_distances, cluster_list, init_score):
        
        self.ven = []
        self.ven_target_pos = []
        self.ven_bikes_to_take = []
        self.ven_start_pos = ven_start_pos
        self.current_time = current_time
        self.cluster_pairwise_distances = cluster_pairwise_distances
        self.cluster_list = cluster_list
        self.score = init_score
        
        for idx in range(len(self.ven_start_pos)):
            van_target = self.closest_station(ven_start_pos[idx])
            van = Van(self.ven_start_pos[idx], self.current_time, van_target, self.cluster_pairwise_distances, idx, self.cluster_list, self.score)
            self.ven.append(van)
            self.ven_target_pos.append(van_target)
            self.ven_bikes_to_take.append(0) # will change
                
    def closest_station(self, cluster):
            col = self.cluster_pairwise_distances[cluster, :]
            # index of second smallest
            cleaned = np.delete(col, cluster)
            idcs = np.argsort(np.abs(cleaned))
            idx = np.random.choice(idcs[0:5])
            #print(idx)
            return idx
    
    def ven_update_directions(self):
        """given current positions of the vans, compute next positions"""
        ## random strategy
#         self.ven_target_pos = []
#         for idx in range(len(self.ven_start_pos)):
#             self.ven_start_pos[idx] = self.ven[idx].current_cluster
#             self.ven_target_pos.append(self.closest_station(self.ven_start_pos[idx]))
#             # update number of bikes to take/ drop off from, negative == drop off, positive = pick up 
#             self.ven_bikes_to_take[idx] = self.ven_bikes_to_take[idx]
        
        ## close to deads
        radius = 7 # this many closest clusters
        #self.ven_target_pos = []
#         for idx in range(len(self.ven_start_pos)):
#             self.ven_start_pos[idx] = self.ven[idx].current_cluster
#             col = self.cluster_pairwise_distances[self.ven[idx].current_cluster, :]
#             # index of second smallest
#             cleaned = np.delete(col, self.ven[idx].current_cluster)
#             closest = np.argsort(np.abs(cleaned))[0:radius]
            
#             # update number of bikes to take/ drop off from, negative == drop off, positive = pick up 
#             for clus_idx in closest:
#                 current_cluster = self.cluster_list[clus_idx]
#                 is_dead = current_cluster.is_close_to_dead()
#                 if is_dead:
#                     num_bikes_in_cluster = current_cluster.num_bikes
#                     capacity = current_cluster.capacity
#                     target = int(np.floor(0.5 * capacity))
#                     self.ven_bikes_to_take[idx] = - target + num_bikes_in_cluster
#                     self.ven_target_pos[idx] = clus_idx
#                     break
        
        ## priority matrix
        radius = 7 # this many closest clusters
        #self.ven_target_pos = []
        for idx in range(len(self.ven_start_pos)):
            self.ven_start_pos[idx] = self.ven[idx].current_cluster
            col = self.cluster_pairwise_distances[self.ven[idx].current_cluster, :]
            # index of second smallest
            cleaned = np.delete(col, self.ven[idx].current_cluster)
            closest = np.argsort(np.abs(cleaned))[0:radius]
            
            top_rank = 0
            top_ranked_neighbour = None
            # update number of bikes to take/ drop off from, negative == drop off, positive = pick up 
            for clus_idx in closest:
                current_cluster = self.cluster_list[clus_idx]
                is_dead = current_cluster.is_close_to_dead()
                current_hour = int(self.current_time.hour)
                flow_this_hour = priority_matrix[clus_idx][current_hour]
                predicted_num = flow_this_hour + current_cluster.num_bikes
                
                rank = max(np.abs(current_cluster.capacity - predicted_num), predicted_num)
                if rank > top_rank:
                    top_rank = rank
                    top_ranked_neighbour = clus_idx
                
            capacity = current_cluster.capacity
            target = int(np.floor(0.5 * capacity))
            self.ven_bikes_to_take[idx] = - target + current_cluster.num_bikes
            self.ven_target_pos[idx] = top_ranked_neighbour
                
                
            
            
    def check_status(self, updated_time, current_score):
        #print(self.ven_start_pos, self.ven_target_pos)
        self.score = current_score
        out_list = []
        self.ven_update_directions()
        
        for idx in range(len(self.ven_start_pos)):
            out_list.append(self.ven[idx].check_status(updated_time, self.ven_target_pos[idx], self.ven_bikes_to_take[idx], self.score))
        return out_list    
        
        #print(self.ven_start_pos, self.ven_target_pos)
        
    
    def update_previous(self):
        for idx in range(len(self.ven_start_pos)):
            self.ven[idx].update_previous(self.ven_start_pos[idx])
        
```

```python
class Van:
    
    speed = 0.0044704 # km / s
    bike_capacity = 40
    current_num_bikes = 15
    previous_cluster = None
        
    def __init__(self, current_cluster, time_of_departure, destination_cluster, cluster_pairwise_distances, name, cluster_list, init_score):
        self.current_cluster = current_cluster
        self.time_departure = time_of_departure
        self.destination_cluster = destination_cluster
        self.cluster_pairwise_distances = cluster_pairwise_distances
        self.name = name
        self.cluster_list = cluster_list
        self.score = init_score
    
    def update_previous(self, prev):
         self.previous_cluster = prev
    
    def update_destination(self, new_destination, current_time):
        self.current_cluster = self.destination_cluster
        self.destination_cluster = None
        self.time_departure = current_time
        
    def check_status(self, current_time, new_destination, num_bikes_to_take, current_score):
    #print(self.current_cluster, self.destination_cluster)
        if new_destination == self.destination_cluster: #still moving
            time_to_destination = self.cluster_pairwise_distances[self.destination_cluster][self.current_cluster] / self.speed
            time_to_destination = time_to_destination + (7 * 60) # load time
            time_to_destination = datetime.timedelta(seconds=time_to_destination)
            if abs(current_time - self.time_departure) >= abs(time_to_destination): # arrived!
                init_cluster = self.current_cluster
                init_time = self.time_departure
                self.update_destination(new_destination, current_time)
                print('\nVan ' + str(self.name) + ' has moved to cluster ' + str(self.current_cluster) + '\n')
                self.load_bikes(num_bikes_to_take)
                self.update_score(current_score)
                tuple_out = (init_time, current_time, self.name, self.previous_cluster, new_destination, self.score)
                print(tuple_out)
                return(tuple_out)
            #else:
                #print('\nVan ' + str(self.name) + ' has not reached cluster ' + str(new_destination) + '\n')
        else: # kick off
            self.destination_cluster = new_destination


    def load_bikes(self, n):
        """pick up / unload n bikes from current cluster"""
        self.current_num_bikes = self.current_num_bikes + n
        if (self.current_num_bikes < 0):
            num_drop_off = -self.current_num_bikes
            for _ in range(num_drop_off):
                self.cluster_list[self.current_cluster].add_bike()
            print('\n' + str(num_drop_off) + ' bikes added to cluster ' + str(self.current_cluster) + ' by van ' + str(self.name) + '\n')
            print('\n Van ' + str(self.name) + ' is full!\n')
            
        if (self.current_num_bikes > self.bike_capacity):
            num_pick_up = self.current_num_bikes - self.bike_capacity
            for _ in range(num_pick_up):
                self.cluster_list[self.current_cluster].remove_bike()
            print('\n' + str(num_pick_up) + ' bikes taken from cluster ' + str(self.current_cluster) + ' by van ' + str(self.name) + '\n')
            print('\n Van ' + str(self.name) + ' is empty!\n')        
        
        if n > 0:
            for _ in range(n):
                self.cluster_list[self.current_cluster].remove_bike()
            print('\n' + str(n) + ' bikes taken from cluster ' + str(self.current_cluster) + ' by van ' + str(self.name) + '\n')
        
        if n < 0:
            for _ in range(-n):
                self.cluster_list[self.current_cluster].add_bike()
            print('\n' + str(-n) + ' bikes added to cluster ' + str(self.current_cluster) + ' by van ' + str(self.name) + '\n')
        
    def update_score(self, current_score):
        self.score = current_score
        
```

```python
class Simulation:
    
    simulation_record = []
    
    score = 0
    
    def initialise_clusters(self):
        self.cluster_list = []
        for index, row in self.cluster_locs.iterrows():
            cluster_id, name, capacity, lat, lon = index, row['name'], row['capacity'], row['lat'], row['lon']
            next_cluster = Cluster(cluster_id, name, capacity, lat, lon)
            self.cluster_list.append(next_cluster)
    
    def initialise_bikes(self):
        ## for now just initialise with frac times the possible bikes
        if type(self.initial_bike_dist) == float:
            frac = self.initial_bike_dist
            for cluster in self.cluster_list:
                for _ in range(int(np.floor(cluster.capacity * frac))):
                    cluster.add_bike() 
        else:
            for counter, cluster in enumerate(self.cluster_list):
                for _ in range(initial_bike_dist[counter]):
                    cluster.add_bike()

    def cluster_population_list(self):
        populations = []
        for cluster in self.cluster_list:
            populations.append(cluster.num_bikes)
        return populations
    
            
    def __init__(self, start_idx, usage_data, cluster_locs, cluster_pairwise_dist, id_to_cluster_dict, initial_bike_dist):
        self.start_idx = start_idx
        self.usage_data = usage_data.sort_values('Start Date')
        self.cluster_locs = cluster_locs
        self.cluster_pairwise_dist = cluster_pairwise_dist
        self.id_to_cluster_dict = id_to_cluster_dict
        self.initial_bike_dist = initial_bike_dist
        self.initialise_clusters()
        self.initialise_bikes()
        
    
    def run_simulation(self, T):
        """run the simulation"""
        
        if (T >= self.usage_data.shape[0]):
            raise(ValueError('too many simulation steps'))
                        
        print("Simulation running from time " + str(self.usage_data.iloc[self.start_idx]['Start Date']) + ' to ' \
              + str(self.usage_data.iloc[self.start_idx + T - 1]['Start Date']) + '\n')
        
        departed = []
         
        
        for t in range(T):
            
            num_clusters = self.cluster_locs.shape[0]
            
            idx = self.start_idx + t
            row = self.usage_data.iloc[idx]
            current_date = row['Start Date']  
            start_id = int(row['StartStation Id'])
            start_cluster = self.id_to_cluster_dict[start_id]
            
            # journey start
            try:
                self.cluster_list[start_cluster].remove_bike()
                departed.append(idx)
            except ValueError:
                self.score = self.score - 1
            
            # search through departed journeys
                      
            print('cluster ' + str(start_cluster) + ' has lost a bike from entry ' + str(idx) + ' at time ' + str(current_date))
            returned = []
            for counter, entry_idx in enumerate(departed):
                if self.usage_data.iloc[entry_idx]['Start Date'] < current_date:
                    returned.append(counter)
                    rrow = self.usage_data.iloc[entry_idx]
                    end_id = int(rrow['StartStation Id'])
                    end_cluster = self.id_to_cluster_dict[end_id]
                    try:
                        self.cluster_list[end_cluster].add_bike()
                        print('cluster ' + str(end_cluster) + ' has gained a bike from entry ' + str(entry_idx) + ' at time ' + str(current_date))
                    except ValueError:
                        self.score = self.score - 1

                    
            # remove returned from departed
            for counter in sorted(returned, reverse=True):
                del departed[counter]
            
            # men with ven
            
            if t == 0:
                ven_start_pos = np.random.choice(num_clusters, 30)
                men_with_ven = MenWithVen(ven_start_pos, current_date, self.cluster_pairwise_dist, self.cluster_list, self.score)
            else:
                try:
                    men_with_ven.update_previous()
                    self.simulation_record = self.simulation_record + men_with_ven.check_status(current_date, self.score)
                except(ValueError):
                    self.score = self.score - 1
            
            #print(current_date, str(self.cluster_population_list()))
            print('current score: ' + str(self.score))
        
        self.simulation_record =  [x for x in self.simulation_record if x is not None]
        
        return self.simulation_record
```

```python
def date_filter(dataset, lower_year, lower_month, lower_day, lower_hour, upper_year, upper_month, upper_day, upper_hour,):
    dataset = dataset[(dataset["Start Date"] >= datetime.datetime(lower_year, lower_month, lower_day, lower_hour))\
                      & (dataset["Start Date"] < datetime.datetime(upper_year, upper_month, upper_day, upper_hour))]
    return dataset
```
